```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Sample data (You can replace this with your own dataset)
data = {
    "Region": ["North", "South", "East", "West", "North", "South", "East", "West", "North", "East"],
    "Item": ["Item1", "Item2", "Item3", "Item1", "Item2", "Item3", "Item1", "Item2", "Item3", "Item1"],
    "Order Date": pd.to_datetime(["2025-04-01", "2025-04-02", "2025-04-03", "2025-04-04", "2025-04-05",
                                  "2025-04-06", "2025-04-07", "2025-04-08", "2025-04-09", "2025-04-10"]),
}

df = pd.DataFrame(data)

# Function to calculate mean, median, and mode for selected column
def calculate_stats(column_name):
    if column_name == "Region" or column_name == "Item":
        messagebox.showerror("Error", "Mean, Median, Mode are not applicable to categorical data like Region or Item.")
        return None, None, None
    column_data = df[column_name]
   
    # For numerical data, calculate the mean, median, and mode
    mean_val = np.mean(column_data)
    median_val = np.median(column_data)
    mode_val = stats.mode(column_data)[0][0]
   
    return mean_val, median_val, mode_val

# Function to display stats and plot the histogram
def display_stats():
    column_name = combo_box.get()
   
    if column_name == "Select":
        messagebox.showerror("Error", "Please select a valid column.")
        return
   
    mean_val, median_val, mode_val = calculate_stats(column_name)
   
    if mean_val is not None:
        label_mean.config(text=f"Mean: {mean_val}")
        label_median.config(text=f"Median: {median_val}")
        label_mode.config(text=f"Mode: {mode_val}")
       
        # Plotting the data
        fig, ax = plt.subplots()
        ax.hist(df[column_name], bins=10, alpha=0.7, color='blue')
        ax.axvline(mean_val, color='r', linestyle='dashed', linewidth=2, label='Mean')
        ax.axvline(median_val, color='g', linestyle='dashed', linewidth=2, label='Median')
        ax.axvline(mode_val, color='y', linestyle='dashed', linewidth=2, label='Mode')
       
        ax.set_title(f'Distribution of {column_name}')
        ax.set_xlabel(column_name)
        ax.set_ylabel('Frequency')
        ax.legend()
       
        # Embed the Matplotlib plot in Tkinter window
        canvas = FigureCanvasTkAgg(fig, master=window)
        canvas.draw()
        canvas.get_tk_widget().grid(row=4, column=0, columnspan=3)

# Create the main window
window = tk.Tk()
window.title("Data Statistics UI")

# Dropdown to select the column
label_select_column = tk.Label(window, text="Select a column:")
label_select_column.grid(row=0, column=0)

combo_box = tk.StringVar(window)
combo_box.set("Select")
dropdown = tk.OptionMenu(window, combo_box, "Region", "Item", "Order Date")
dropdown.grid(row=0, column=1)

# Labels to display Mean, Median, Mode
label_mean = tk.Label(window, text="Mean: ")
label_mean.grid(row=1, column=0, columnspan=2)

label_median = tk.Label(window, text="Median: ")
label_median.grid(row=2, column=0, columnspan=2)

label_mode = tk.Label(window, text="Mode: ")
label_mode.grid(row=3, column=0, columnspan=2)

# Button to calculate stats
button_calculate = tk.Button(window, text="Calculate", command=display_stats)
button_calculate.grid(row=0, column=2)

# Start the Tkinter event loop
window.mainloop()
```
